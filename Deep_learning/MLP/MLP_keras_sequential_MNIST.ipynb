# Implementing MLPs with Keras
## Building an Image Classifier Using the Sequential API
### Using Keras to load the dataset

```python
mnist = tf.keras.datasets.mnist

# PARTICION EXTERNA
(X_train, y_train), (X_test, y_test) = mnist.load_data()
print('TRAINING DATA')
print(X_train.shape)
print(y_train.shape)
print('\nTESTING DATA')
print(X_test.shape)
print(y_test.shape)

# el shape es [cantidad de imagenes, pixesls en x, pixles en y]
```

### Inspecting the dataset

```python
import matplotlib.pyplot as plt

def display_digit(num, X, y):
    image = X[num, :, :] 
    label = y[num]

    plt.imshow(image, 
               cmap=plt.get_cmap('gray_r'))
    plt.title('Example %d, label %d' % (num, label))
    plt.show()
    return None

for index in [i for i in np.random.randint(0, X_train.shape[0], size= 3)]:  # voy a plotear 3 numeros aleatorios
    display_digit(index, X_train, y_train)
```

### data preprocesing

```python
from tensorflow.keras.utils import to_categorical
from sklearn.model_selection import train_test_split


X_train, X_test = X_train/ 255.0, X_test/ 255.0         # normalización de los datos entre 0 y 1
y_train = to_categorical(y_train, num_classes= 10)       # One hot encode de las etiquetas
y_test = to_categorical(y_test, num_classes= 10)

print('TARGETS ONE HOT ENCODE')
print(y_train.shape, y_test.shape)
print(f'new one hot label: {y_train[5000]} wich label is: {y_train[5000].argmax(axis= 0)} \n')

# PARTICION INTERNA
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.1, random_state=42)
print(X_train.shape, y_train.shape)
print(X_val.shape, y_val.shape)
```

### Creting the model using a sequential API

#### CREATE

```python
from tensorflow.keras.models import Sequential          # estructura sequencial
from tensorflow.keras.layers import Dense, Flatten      # dos tipos de capas

model = Sequential()                        # create
model.add(Flatten())                        # la capa input; en un MLP debe ser un vector 1D
model.add(Dense(512,                        # la capa oculta: 512 neuronas
                input_shape= (28*28,),      # a la primera capa oculta se le debe indicar el input shape
                activation= 'relu'))        # la función de activación
model.add(Dense(10,                         # capa de salida: tantas neuronas como clases a predecir
                activation= 'softmax'))     # la funcion de activación para clasificación multiclase
```

#### COMPILE

```python
from tensorflow.keras.optimizers import SGD

model.compile(loss= 'categorical_crossentropy',     # esto en caso de que las etiquetas esten en one hot
              optimizer= SGD(0.005),                 # el optimizador que vamos a usar con su learning rate
              metrics= ['accuracy'])                # la metrica que queremos observar
```

```python
# print(model.summary())
```

#### fit ---> TRAIN

```python
from time import time
```

```python
start = time()

print('INFO: entrenando red neuronal...')

# El modelo guarda en cada epoca las perdidas y el acuracy en cada epoca
# el 422 son los steps es decir son batch sizes de 422 imagenes en cada epoca

H = model.fit(X_train, y_train, 
              validation_data= (X_val, y_val), 
              epochs= 50, 
              batch_size= 128)
end = time()
train_time = end - start
print(f'El tiempo de entrenamiento a sido de {train_time} segundos')
```

```python
print(model.summary())
```

#### Learning curves

```python
plt.style.use('ggplot')
plt.figure(figsize= (10, 7))
plt.plot(np.arange(0, 50), H.history['loss'], label = 'train_loss')
plt.plot(np.arange(0, 50), H.history['val_loss'], label= 'validation_loss')
plt.plot(np.arange(0, 50), H.history['accuracy'], label= 'training accuracy')
plt.plot(np.arange(0, 50), H.history['val_accuracy'], label = 'val accuracy')
plt.title('Training - validation Loss and accuracy ')
plt.xlabel('epochs')
plt.ylabel('loss and accuracy')
plt.legend(loc= 'best')
plt.show()
```

#### predict ---> evaluate the model

```python

from sklearn.metrics import classification_report
print('[INFO]: Evaluando la red neuronal ...')
predictions = model.predict(X_test, batch_size= 128)
print('\nLets predict an example:')
print(f'label: {y_test[0].argmax()}, prediction: {predictions[0].argmax()}\n')
print(classification_report(y_test.argmax(axis= 1), predictions.argmax(axis= 1)))
```
