# Filter

```python
import numpy as np
import pandas as pd
```

## cargando df

```python
path = 'http://archive.ics.uci.edu/ml/machine-learning-databases/auto-mpg/auto-mpg.data'

df = pd.read_csv(path, delim_whitespace=True, header=None,
            names = ['mpg', 'cilindros', 'desplazamiento','potencia',
            'peso', 'aceleracion', 'año', 'origen', 'nombre'],
            na_values='?')

df.sample(10)
```

## filter con lambda

```python
filtro = filter(lambda x: 'ford' in x, df['nombre'].str.lower())
list(filtro)

# dame los nombres de los coches que contengan ford
```

## filter con funciones personalizadas

```python
def media_de_potencia_mas_100(x):
    return x['potencia'].mean() > 100

display(df.head())
```

```python
df.groupby('nombre').filter(media_de_potencia_mas_100)

# dame todas las columnas agrupadas por nombre que tengan una potencia media superior a 100
```
