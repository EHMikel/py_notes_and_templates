# .Groupby()

## sintaxis básica

dataframe.**groupby**('agrupamelas por estas columnas')[[lista de columnas que quiero que agrupes]].funcion_agregada())


```python
import numpy as np 
import pandas as pd
```

## cargando dataframes

### dataframe realista

```python
path = 'http://archive.ics.uci.edu/ml/machine-learning-databases/auto-mpg/auto-mpg.data'

df = pd.read_csv(path, delim_whitespace=True, header=None,
            names = ['mpg', 'cilindros', 'desplazamiento','potencia',
            'peso', 'aceleracion', 'año', 'origen', 'nombre'],
            na_values='?')

df.sample(5)
```

## El groupby mas básico

**NOTA**: si vamos a aplicar una función puramente númerica como mean() o median() es mejor seleccionar las columnas por las que queremos que lo calcule

### dataframe realista

```python
df.groupby('nombre').max()
```

## El truco de loc para groupby

**NOTA**: Obligatorio incluir en el loc la columna por la queremos que agrupe, en este caso género 



### dataframe realista

```python
mask = df['potencia'] > 100

df.loc[mask,['nombre', 'potencia', 'peso', 'aceleracion']].groupby('nombre').median()
```

## Aggregate([lista de fuunciones] en groupby()

### Dataframe realista

```python
df.head()
```

```python
df.groupby('nombre')[['potencia', 'peso', 'aceleracion']].aggregate([max, min, np.mean]).head()

# agrupame por nombre, las columnas potencia peso y aceleracion, y aplicale a cada una max min y mean
```

```python
df.groupby('nombre')[['potencia', 'peso', 'aceleracion']].aggregate([max, min, np.mean]).columns
```

## FORMA RECOMENDADA: Aggregate --> groupby()

### dataframe realista

```python
df.groupby('nombre').aggregate(
{'potencia':np.mean,
 'peso':'max',
 'aceleracion': np.median
    
}).head()
```

## Forma recomendada 2 groupby 

 ### dataframe realista

```python
df.groupby('nombre').aggregate(
                    maximo_potencia= pd.NamedAgg(column= 'potencia', aggfunc='max'),
                    potencia_media= pd.NamedAgg(column='potencia', aggfunc='mean'))
```
