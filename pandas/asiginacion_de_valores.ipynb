```python
import numpy as np
import pandas as pd
```

```python
path = 'http://archive.ics.uci.edu/ml/machine-learning-databases/auto-mpg/auto-mpg.data'

df = pd.read_csv(path, delim_whitespace=True, header=None,
            names = ['mpg', 'cilindros', 'desplazamiento','potencia',
            'peso', 'aceleracion', 'año', 'origen', 'nombre'],
            na_values='?')

display(df.sample(5))
```

# como asignar valores correctamente en pandas

## metodo condicional: INCORRECTO

```python
mask = df['potencia'] < 100
```

```python
# df[mask]= 0 # esto se lo aplica a toda la fila, no a la columna que queremos
```

```python
df.sample(3)
```

##  metodo loc[row, col] = : CORRECTO

```python
mask = df['potencia'] < 100
```

```python
df.loc[mask,'potencia'] = np.nan # a los qeu tengan potencia inferior a 100 asignales un nan
```

```python
df.sample(5)
```

## metodo np.where(condicion, si se cumple, si no se cumple): CORRECTO

```python
#condicion = df['potencia'] == np.nan

condicion = pd.isnull(df['potencia'])
```

```python
%%timeit
df['potencia'] = np.where(condicion, df['potencia'].mean(),  df['potencia'])
```

```python
df.sample(5)
```

## metodo apply(): CORRECTO

ES mas eficiente que where

```python
def subir_10(serie):
    if serie < 150:
        serie+=10
    return serie
```

```python
%%timeit
df['potencia'] = df['potencia'].apply(subir_10)
```

```python
df.sample(5)
```
