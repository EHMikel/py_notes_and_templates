```python
import numpy as np
import pandas as pd
import os
```

```python
path= 'C:\/Users/plane/OneDrive/Escritorio/COMPUTING SCIENCE/MASTER Inteligencia artificial/python para IA/S11 data procesing/res/ml-1m'

ruta_users = os.path.join(path, "users.dat")
ruta_ratings = os.path.join(path, "ratings.dat")
ruta_movies = os.path.join(path, "movies.dat")

users_df = pd.read_csv(ruta_users, sep='::', index_col=0,
                       header=None, names=['UserID','Gender','Age','Occupation','Zip-code'],
                       engine='python', encoding="ISO-8859-1")

ratings_df = pd.read_csv(ruta_ratings, sep='::', index_col=0, 
                        header=None, names=['UserID','MovieID','Rating','Timestamp'],
                        engine='python', encoding="ISO-8859-1")

movies_df = pd.read_csv(ruta_movies, sep='::', index_col=0, 
                        header=None, names=['MovieID','Title','Genre'],
                        engine='python', encoding="ISO-8859-1")
```

 # Uniendo datasets con 'join' y 'merge'
- merge() == join()
- 'join' utiliza por defecto los índices para unir
- Utilizando el parámetro 'on'
 - Si las columnas difieren, 'left_on' y 'right_on'
- Diagrama de Venn para unión de tablas:

```python
users_df.shape, ratings_df.shape # no comparten numero de columnas ni filas
```

## metodo on = 'nombre_columna'

```python
combined_df = users_df.merge(right=ratings_df, on='UserID', how='inner') # parametro 'on' define la columna pivote
                                                                         # tienen la columa userID en común, solo que
                                                                         # cada user a rateado más de una pelicula

display(combined_df.head(5))
print(len(combined_df))
```

## metodo right_index= True, left_index= True

```python
combined_df = users_df.merge(right=ratings_df,
                                       left_index=True,
                                       right_index=True,
                                       how='inner') # parametro 'on' define la columna pivote


display(combined_df.head(5))
print(len(combined_df))
```
