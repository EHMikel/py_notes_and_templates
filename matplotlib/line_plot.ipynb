# line plot

![](https://matplotlib.org/3.1.0/_images/anatomy.png)


Múltiples parámetros

```python
plt.plot(x, y, style, label, linewidth=1, alpha=1)
```

- 'style': cadena de texto con formato (color, tipo...) \[color\]\[marker\]\[line\]
- 'label': referencia para leyenda
- 'linewidth': brocha
- 'alpha': transparencia

Otros: https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html

## line plot básico

````python
x = list(range(10)) # Datos del eje X
y = list(np.random.randint(-10,10, size= (10))) # Datos del eje Y


fig = plt.figure(figsize=(7,4))
# Función principal:

plt.plot(x,              # Datos del eje X
         y,              # Datos del eje Y
         label='Línea',  # Nombre de la línea
         color="blue")   # Color de la línea

# Añadiendo título
plt.title('Una línea')

# Etiqueta de los ejes
plt.xlabel('El eje de las X')
plt.ylabel('El eje de las Y')

# Añadiendo una leyenda
plt.legend(loc='lower right') # tambien podria ser 'lower left'

# plt.savefig('path/nombre_figura.png') # para guardar la figura

# Mostrando la figura
plt.show()
````

### se puede meter en una función

````python
def grafica_linea(x_axis, y_axis):    
    
    plt.plot(x_axis, 
             y_axis,
             label='Línea',
             color="blue")
    
    plt.title('Una línea')
    
    plt.xlabel('El eje de las X')
    plt.ylabel('El eje de las Y')
    
    plt.legend(loc='lower right')
    
    plt.show()
````

````python
grafica_linea(x, y)
````

## Varias lineas en un mismo gráfico

````python
x = np.arange(30)
y0 = np.random.randint(30,80, size= (30))
y1 = np.random.randint(20,50, size= (30))
y2 = np.random.randint(70,85, size= (30))
#y3 = np.random.randint(5,30 size= (100))


fig = plt.figure(figsize= (10, 6))

plt.plot(x, y0,                                        # datos x e y
         'go-',                                        # el color de la linea
         label= 'green', linewidth= 1, alpha= 0.5)     # etiqueta, grosor de linea, opacidad

plt.plot(x, y1,
         'bx-',
         label= 'blue', linewidth= 1, alpha= 0.9)

plt.plot(x, y2,
         'rh-',
         label= 'red', linewidth= 1, alpha= 0.7)

plt.legend(loc= 'lower left')                                  # muestra la legenda y le paso la localizacion
plt.title('3 lineas en un grafico', fontdict={'fontsize': 16}) # el titulo y el tamaño del mismo
plt.ylabel('valores aleatorios')
plt.xlabel('valores continuos de x')
# plt.grid() # para mostrar la cuadrícula


plt.show()
````

## subgraficos 

### Decoradores

Cambian el aspecto general de la figura
```python
axes.set_xticks(<lista>) #etiquetas del eje
axes.set_title(‘titulo’) #titulo
axes.set_ylabel(‘label’)
axes.set_frame_on(bool)
```
[Más métodos](https://matplotlib.org/stable/api/axes_api.html)


````python
x = np.arange(30)
y0 = np.random.randint(30,80, size= (30))
y1 = np.random.randint(20,50, size= (30))
y2 = np.random.randint(70,85, size= (30))

#fig = plt.figure()
fig, ax = plt.subplots(3, 1, figsize= (15,10)) # cuando usamos subplots el figsize se pone aqui
                                             # plt.subplots(row, col, figsize)
plt.title('3 linas en 3 graficos', loc='center')

ax[0].plot(x,y0,
        'go-',                                        # el color de la linea
         label= 'green', linewidth= 1, alpha= 0.5)
ax[0].set_title('verde')


ax[1].plot(x, y1,
         'bx-',
         label= 'blue', linewidth= 1, alpha= 0.9)
ax[1].set_title('blue')



ax[2].plot(x, y2,
         'rh-',
         label= 'red', linewidth= 1, alpha= 0.7)
ax[2].set_title('rojo')

plt.legend()

plt.show()
````

## Ecuaciones matemáticas

````python
x = np.arange(-3,4)

fig = plt.figure(figsize= (5,5))

plt.plot(x, x, color= 'red', label= 'linear')
plt.plot(x, x**2-2, color = 'blue', label= 'cuadratic')
plt.plot(x, x**3-3, color = 'purple', label= 'cubic')

plt.grid()
plt.legend()
plt.show()
````

## Anotaciones

Formas de añadir información sobre el gráfico

```python
axes.text(x,y,’texto’)
axes.annotate(texto, xy, xytext, xycoords, arrowprops)
```

[Más métodos](https://matplotlib.org/stable/tutorials/text/annotations.html)

````python
y_data = np.random.randint(100, size=(10)) # Cuidado con las dimensiones absolutas
x_data = np.arange(10)

fig,axis = plt.subplots(1,1, figsize= (7,5))
axis.plot(x_data,y_data)
axis.text(x_data[3],y_data[3],'Low income') # coordenadas igual a los datos
axis.annotate('Low Income', xy=(x_data[8],y_data[8]), xytext=(4, 10), arrowprops = dict(facecolor='red', shrink=0.5)) # anotación completa
plt.show() 
````
